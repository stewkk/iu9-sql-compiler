```python
import json

data = json.loads(data)    
```

```python
import re
from typing import Any

benchmarks: list[dict[str, Any]] = list()

for benchmark in data['benchmarks']:
    match = re.match('BM_SQL<(.*), (.*)>', benchmark['name'])
    if match is None:
        raise ValueError("invalid benchmark name")
    executor, query = match.group(1), match.group(2)
    real_time, cpu_time = benchmark['real_time'], benchmark['cpu_time']
    time_unit = benchmark['time_unit']
    if time_unit != 'ns':
        raise ValueError("time_unit is not ns")
    benchmarks.append({
        'executor': executor,
        'query': query,
        'real_time': real_time,
        'cpu_time': cpu_time,
    })
    
import pandas as pd

benchmarks = pd.DataFrame(benchmarks)

benchmarks
```

```python
benchmarks = benchmarks[benchmarks['query'].str.contains('Complex')].copy()
benchmarks['query'] = benchmarks['query'].str.extract(r'(\d+)').astype(int)
benchmarks.sort_values(by='query', inplace=True)
benchmarks.reset_index(drop=True, inplace=True)
benchmarks
```

```python
import matplotlib.pyplot as plt

benchmarks.pivot(index='query', columns='executor', values='cpu_time').plot(kind='bar')
plt.ylabel('CPU time (ns)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
pivoted = benchmarks.pivot(index='query', columns='executor', values='cpu_time')
pivoted['speedup'] = 1.0 - pivoted['CachedJitCompiledExpressionExecutor'] / pivoted['InterpretedExpressionExecutor']
pivoted
```

```python
pivoted['speedup'].plot(kind='bar')
plt.ylabel('speedup (%)')
plt.xlabel('query')
plt.xticks(rotation=0, ha='center')
plt.tight_layout()
plt.show()
```
